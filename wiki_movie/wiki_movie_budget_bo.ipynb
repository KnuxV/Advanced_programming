```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python

# Load the data
df = pd.read_csv('data/wiki_movie_plots_deduped.csv')
```

```python
# Now we generalize the process to get the budget and box office for all movies 


def get_budget_box_office(url):
    r = requests.get(url)
    if r.status_code != 200:
        return np.nan, np.nan
    soup = BeautifulSoup(r.content)
    try:
        th_budget = soup.find("table").find('th', class_='infobox-label', string="Budget")
    except AttributeError:
        return np.nan, np.nan
    if th_budget is None:
        budget = np.nan
    else:
        budget_value = th_budget.find_next('td', class_='infobox-data')
        budget = budget_value.text
    
    try:
        th_box_office = soup.find("table").find('th', class_='infobox-label', string="Box office")
    except AttributeError:
        return np.nan, np.nan
    if th_box_office is None:
        box_office = np.nan
    else:
        box_office_value = th_box_office.find_next('td', class_='infobox-data')
        box_office = box_office_value.text
    return budget, box_office
```

```python
# We do it 50 random movies from 2015
df_american_2010 = df[
    (df['Release Year'] >= 2010 )
     & 
    (df['Origin/Ethnicity'] == 'American')
]

# Apply the get_budget_box_office function and expand the result into separate columns
df_american_2010['Budget'], df_american_2010['Box Office'] = zip(*df_american_2010['Wiki Page'].apply(get_budget_box_office))
```

```python
df_american_2010 = df_american_2010.dropna(subset=['Budget', 'Box Office'])
df_american_2010
```

```python
import pandas as pd
import re

# Updated regex pattern to handle ranges with multiple dollar signs
pattern = r'\$?([\d,.]+)\s?(?:–\$?([\d,.]+))?\s?(million|billion)?'

# Function to extract and clean the numeric values
def extract_value(text):
    # Remove parentheses and content within, as well as any references like [1]
    text = re.sub(r'\(.*?\)|\[.*?\]', '', text)
    
    matches = re.findall(pattern, text)
    
    if not matches:
        return None  # Return None if no matches
    
    extracted_values = []
    
    for match in matches:
        low_str, high_str, scale = match
        
        # Handle ranges like $40–$45
        if high_str:
            low = float(low_str.replace(',', ''))
            high = float(high_str.replace(',', ''))
            number = (low + high) / 2  # Take the average of the range
        else:
            number = float(low_str.replace(',', ''))
        
        # Convert to appropriate scale
        if scale == 'million':
            number *= 1_000_000
        elif scale == 'billion':
            number *= 1_000_000_000
        
        extracted_values.append(number)
    
    # Return the first extracted value (if there are multiple values like gross/net)
    return extracted_values[0] if extracted_values else None

# Apply the function to the 'Budget' column and create a new 'Clean_Budget' column
df_american_2010['Clean_Budget'] = df_american_2010['Budget'].apply(extract_value) / 1_000_000
df_american_2010['Clean_Box_Office'] = df_american_2010['Box Office'].apply(extract_value) / 1_000_000
# Show the DataFrame with cleaned values
df_american_2010
```
